## Problem 7 (27 lines on a cubic)

````julia
using HomotopyContinuation
````

### Part (a): Extracting coefficients

````julia
# Introducing the variables we will use
@var x y z t p[1:3] v[1:3]
````

````julia
# The polynomial defining the surface
f = 81*(x^3+y^3+z^3)-189*(x^2*y+x^2*z+x*y^2+x*z^2+y^2*z+y*z^2)+54*x*y*z+126*(x*y+x*z+y*z)-9*(x^2+y^2+z^2)-9*(x+y+z)+1
````

````julia
# Parametrization of line
parametrized_line = p+t*v
````

````julia
line_evaluated_at_surface = subs(f, [x;y;z] => parametrized_line)
````

````julia
# We extract the coefficients c0, c1, c2, c3 as polynomials in p and v
coeffs = coefficients(line_evaluated_at_surface,[t])
````

### Part (b): Solving the system and certifying the solutions

````julia
p_hyperplane = 7 + p[1] + 3*p[2] + 5*p[3]
#p_hyperplane = sum(p.*randn(3))+randn()
````

````julia
v_hyperplane = 11 + 3*v[1] + 5*v[2] + 7*v[3]
#v_hyperplane = sum(v.*randn(3))+randn()
````

````julia
F = System(vcat(coeffs,p_hyperplane,v_hyperplane), variables=vcat(p,v))
````

````julia
res = solve(F, start_system = :total_degree)
````

````julia
sol = solutions(res)
````

````julia
cert_res = certify(F,res)
````

### Part (c): Visualizing the lines
