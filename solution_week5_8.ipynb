### Part (b): Solve the moment system

```julia
@var a b v w lambda

moments = [lambda*a+(1-lambda)*b,
lambda*(a^2+v)+(1-lambda)*(b^2+w),
lambda*(a^3+3*a*v)+(1-lambda)*(b^3+3*b*w),
lambda*(a^4+6*a^2*v+3*v^2)+(1-lambda)*(b^4+6*b^2*w+3*w^2),
15*lambda*v^2*a + 10*lambda*v*a^3 + lambda*a^5 + 15*(1-lambda)*w^2*b + 10*(1-lambda)*w*b^3 + (1-lambda)*b^5]

F = System(moments-sample_moments, variables =[a,b,v,w,lambda]);
```

```julia
solution_result = solve(F, start_system = :total_degree)
```

```julia
list_of_solutions = solutions(solution_result)
```

### Part (c): Certify and check realness

```julia
certification_result = certify(F, list_of_solutions)
```

We have **four** real solutions!

Here is a way to extract the verifiably real solutions from the certification result. (Alternatively, if one trusts the guesses from the `solve` command about which solutions are real, one can use the `real_solutions` command.)

**Note:** All of the real solutions turn out to make statistical sense, in the sense that they are positive, with $\lambda\in(0,1)$.

```julia
list_of_real_solutions = [ real(c.solution) for c in certificates(certification_result) if is_real(c) ]
```

**Note:** We have a symmetry $(a,b,u,v,\lambda)\mapsto (b,a,v,u,1-\lambda)$. Up to this, there are only **two** meaningfully different statistically significant solutions.

### Part (d): Pick the best solution

**Strategy:** Plug in the solutions in $m_6$, and see which value is the closest to $\widehat{m}_6$.

```julia
m6 = 15*lambda*v^3 + 15*(1-lambda)*w^3 + 45*lambda*v^2*a^2 + 15*lambda*v*a^4 + lambda*a^6 + 45*(1-lambda)*w^2*b^2 + 15*(1-lambda)*w*b^4 + (1-lambda)*b^6 ;
```

```julia
m6_sample = sum(data.^6)/length(data)
```

```julia
test_function  = s -> abs(evaluate(m6,[a,b,v,w,lambda]=>s) - m6_sample)
```

```julia
m6_evaluated = map(test_function, list_of_real_solutions)
```

```julia
best_index = findmin(m6_evaluated)[2]
best_solution = list_of_real_solutions[best_index]
```

A slightly smoother way of doing this is to use the `argmin` command in Julia.

```julia
best_solution = argmin(test_function,list_of_real_solutions)
```

### Part (e): Plotting the probability density function

```julia
mix1 = Normal(best_solution[1], sqrt(best_solution[3]));
mix2 = Normal(best_solution[2], sqrt(best_solution[4]));
mixture = MixtureModel(Normal[mix1,mix2],[best_solution[5],1-best_solution[5]]);
```

### Some history

When Karl Pearson introduced his method of moments in the paper [Contributions to the mathematical theory of evolution](https://royalsocietypublishing.org/doi/10.1098/rsta.1894.0003) (Phil. Trans. Roy. Soc., 1894), it involved solving the system above **by hand** (by first reducing it to a univariate polynomial of degree 9).

Solving this was a tremendous amount of work, which made everyone very skeptical about the future promise of his method. Pearson himself wrote in the that

> it may be questioned whether the general theory could ever be applied in practice to any numerical case

and his colleague Charlier wrote that

> Mr. Pearson has indeed possessed the energy to perform this heroic task in some instances [...] but I fear he will have few successors.

Little did they know that 129 years later, a bunch of MSc students in Copenhagen would be able to replicate Pearson's analysis in a 3-hour exercise class!
